```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import (
    mean_squared_error, r2_score, confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
)
from scipy.stats import zscore

# If using Jupyter, you can uncomment the following line:
%matplotlib inline

# =========================
# User-Adjustable Parameters
# =========================
confidence_level = 99.5
Z_value = 4
show_full_lists = False
Outlier_threshold = 20
num_cv_splits = 10
random_state = 42
num_random_states = 1
data_file_path = 'C:/Users/flori/Documents/BEP 2024/Test/Filtered Material Data for BEP - Huisman _Retest Project.xlsx'
output_folder = os.path.dirname(data_file_path)
minimum_sample_size = 50

# Model parameters
rf_n_estimators = 250
rf_max_depth = 50
rf_min_samples_split = 5
rf_min_samples_leaf = 1
rf_max_features = 'sqrt'
rf_bootstrap = True
rf_oob_score = False

et_n_estimators = 250
et_max_depth = 50
et_min_samples_split = 2
et_min_samples_leaf = 2
et_max_features = 'sqrt'

def get_mechanical_requirements(grade, t):
    grade = str(grade).strip().upper()
    if grade == 'S690':
        if t <= 50:
            return {'min_yield_strength': 690, 'tensile_range': (770, 940), 'min_elongation': 10}
        elif t <= 100:
            return {'min_yield_strength': 650, 'tensile_range': (760, 930), 'min_elongation': 10}
        elif t <= 150:
            return {'min_yield_strength': 630, 'tensile_range': (710, 900), 'min_elongation': 10}
        else:
            return {'min_yield_strength': 630, 'tensile_range': (710, 900), 'min_elongation': 10}
    elif grade == 'S355' or 'Q345':
        if t <= 16:
            return {'min_yield_strength': 355, 'tensile_range': (470, 630), 'min_elongation': 22}
        elif t <= 40:
            return {'min_yield_strength': 345, 'tensile_range': (470, 630), 'min_elongation': 22}
        elif t <= 63:
            return {'min_yield_strength': 335, 'tensile_range': (470, 630), 'min_elongation': 22}
        elif t <= 80:
            return {'min_yield_strength': 325, 'tensile_range': (470, 630), 'min_elongation': 21}
        elif t <= 100:
            return {'min_yield_strength': 315, 'tensile_range': (470, 630), 'min_elongation': 21}
        elif t <= 150:
            return {'min_yield_strength': 295, 'tensile_range': (450, 600), 'min_elongation': 21}
        elif t <= 200:
            return {'min_yield_strength': 285, 'tensile_range': (450, 600), 'min_elongation': 21}
        elif t <= 250:
            return {'min_yield_strength': 275, 'tensile_range': (450, 600), 'min_elongation': 21}
        elif t <= 400:
            return {'min_yield_strength': 265, 'tensile_range': (450, 600), 'min_elongation': 19}
        else:
            return {'min_yield_strength': 265, 'tensile_range': (450, 600), 'min_elongation': 19}
    else:
        # default fallback to S690
        if t <= 50:
            return {'min_yield_strength': 690, 'tensile_range': (770, 940), 'min_elongation': 10}
        elif t <= 100:
            return {'min_yield_strength': 650, 'tensile_range': (760, 930), 'min_elongation': 10}
        elif t <= 150:
            return {'min_yield_strength': 630, 'tensile_range': (710, 900), 'min_elongation': 10}
        else:
            return {'min_yield_strength': 630, 'tensile_range': (710, 900), 'min_elongation': 10}

def clean_numeric(value):
    if isinstance(value, str):
        value = value.strip()
        if '/' in value:
            try:
                numbers = [float(num) for num in value.split('/')]
                result = sum(numbers) / len(numbers)
                return result
            except ValueError:
                return np.nan
        value = re.sub(r'[^\d.-]', '', value)
        try:
            return float(value)
        except ValueError:
            return np.nan
    return pd.to_numeric(value, errors='coerce')

def create_enhanced_features(df):
    enhanced = df.copy()
    enhanced['Yield_Ductility_Product'] = enhanced['QC_Re'] * enhanced['QC_A']
    enhanced['Work_Hardening_Range'] = enhanced['QC_Rm'] - enhanced['QC_Re']
    enhanced['Normalized_Work_Hardening'] = (enhanced['QC_Rm'] - enhanced['QC_Re']) / enhanced['QC_Re']
    enhanced['Work_Hardening_Ductility'] = (enhanced['QC_Rm'] - enhanced['QC_Re']) * enhanced['QC_A']
    enhanced['Ductility_Weighted_Strength'] = enhanced['QC_Rm'] * np.sqrt(enhanced['QC_A'])
    enhanced['Quality_Index'] = enhanced['QC_Rm'] + 150 * np.log(enhanced['QC_A'])
    enhanced['Performance_Index'] = enhanced['QC_Re'] * np.power(enhanced['QC_A'], 0.333)
    enhanced['Strain_Hardening_Coefficient'] = np.log(enhanced['QC_Rm']/enhanced['QC_Re']) / np.log(enhanced['QC_A']/100)
    enhanced['Strain_Energy'] = (enhanced['QC_Rm'] + enhanced['QC_Re']) * enhanced['QC_A'] / 2
    enhanced['Strength_Ductility_Balance'] = (enhanced['QC_Rm'] * enhanced['QC_A']) / enhanced['QC_Re']
    enhanced['Formability_Index'] = enhanced['QC_A'] * np.sqrt(enhanced['QC_Rm']/enhanced['QC_Re'])

    enhanced['Thickness_Strength_Index'] = enhanced['QC_Rm'] / np.log(enhanced['Dimension'] + 1)
    enhanced['Thickness_Ductility_Ratio'] = enhanced['QC_A'] / enhanced['Dimension']
    enhanced['Hall_Petch_Approximation'] = enhanced['QC_Re'] * np.sqrt(enhanced['Dimension'])
    enhanced['Size_Effect_Factor'] = enhanced['QC_Re'] * np.power(enhanced['Dimension'], -0.5)
    enhanced['Thickness_Quality_Index'] = enhanced['Quality_Index'] / np.log(enhanced['Dimension'] + 1)
    enhanced['Thickness_Strain_Energy'] = enhanced['Strain_Energy'] / enhanced['Dimension']
    enhanced['Normalized_Thickness_Strength'] = enhanced['QC_Re'] / (enhanced['Dimension'] * enhanced['QC_Rm'])
    enhanced['Thickness_Performance_Factor'] = enhanced['Performance_Index'] / np.sqrt(enhanced['Dimension'])
    enhanced['Surface_Volume_Strength'] = enhanced['QC_Re'] / enhanced['Dimension']

    enhanced['Yield_Tensile_Ratio'] = enhanced['QC_Re'] / enhanced['QC_Rm']
    enhanced['Total_Elongation_Energy'] = enhanced['Strain_Energy'] * enhanced['QC_A']
    enhanced['Strength_Difference_Ratio'] = (enhanced['QC_Rm'] - enhanced['QC_Re']) / enhanced['QC_Rm']
    enhanced['Modulus_of_Toughness'] = 0.5 * (enhanced['QC_Re'] + enhanced['QC_Rm']) * enhanced['QC_A']
    enhanced['Hardening_Coefficient'] = np.log(enhanced['QC_Rm'] / enhanced['QC_Re'])
    enhanced['Elongation_Thickness_Ratio'] = enhanced['QC_A'] / enhanced['Dimension']
    enhanced['Reciprocal_Yield_Strength'] = 1 / enhanced['QC_Re']
    enhanced['Reciprocal_Tensile_Strength'] = 1 / enhanced['QC_Rm']

    return enhanced

def calculate_prediction_intervals_for_tree_model(model, X_test, confidence_level):
    all_tree_preds = np.array([est.predict(X_test) for est in model.estimators_])
    alpha = (100 - confidence_level) / 2
    lower_percentile = alpha
    upper_percentile = 100 - alpha
    y_pred_lower = np.percentile(all_tree_preds, lower_percentile, axis=0)
    y_pred_upper = np.percentile(all_tree_preds, upper_percentile, axis=0)
    return y_pred_lower, y_pred_upper

def process_tests(df, test_columns, direction_column, prefix):
    df = df.copy()
    def calculate_mean_non_zero(row):
        non_zero_values = [value for value in row[test_columns] if value != 0]
        return np.mean(non_zero_values) if non_zero_values else np.nan

    df[f'{prefix}_Mean_Test'] = df.apply(calculate_mean_non_zero, axis=1)

    def check_failure(row):
        mean_value = row[f'{prefix}_Mean_Test']
        threshold = 0.7 * mean_value
        tests = row[test_columns]
        if any(t < threshold for t in tests):
            return 'Fail'
        return 'Pass'

    df[f'{prefix}_Pass_Fail'] = df.apply(check_failure, axis=1)
    df[f'{prefix}_Direction'] = df[direction_column]
    return df

print("Loading and cleaning data...")
df = pd.read_excel(data_file_path, sheet_name=0, na_values=['', ' ', 'NA'])

cert_test_columns = ['QC_V1', 'QC_V2', 'QC_V3']
cert_direction_column = 'QC_Orientation'
real_test_columns = ['Nr1', 'Nr2', 'Nr3']
real_direction_column = 'ImpactOrientation'
info_columns = ['Material Code', 'Manufacturer', 'Yield Strength [Mpa]']

for col in cert_test_columns + real_test_columns + info_columns + ['QC_Re', 'QC_Rm', 'QC_A', 'Dimension', 'Tensile Strength [in MPA]', 'Elongation']:
    df[col] = df[col].apply(clean_numeric)

for col in [cert_direction_column, real_direction_column]:
    df[col] = df[col].astype(str).str.strip().str.upper()

df_clean = df.dropna(subset=cert_test_columns + real_test_columns + [cert_direction_column, real_direction_column]).copy()

print("Processing test results...")
df_processed = process_tests(df_clean, cert_test_columns, cert_direction_column, 'Cert')
df_processed = process_tests(df_processed, real_test_columns, real_direction_column, 'Real')
df_processed = df_processed.copy()

def compare_tests(row):
    if row['Cert_Direction'] != row['Real_Direction']:
        return np.nan
    else:
        return row['Cert_Pass_Fail'] == row['Real_Pass_Fail']

df_processed['Prediction_Correct'] = df_processed.apply(compare_tests, axis=1)
df_matched = df_processed.dropna(subset=['Prediction_Correct'])

accuracy = df_matched['Prediction_Correct'].mean()
print(f"\nOverall Prediction Accuracy (Matching Directions): {accuracy:.2%}")

failed_items = df_matched[
    (df_matched['Cert_Pass_Fail'] == 'Fail') | (df_matched['Real_Pass_Fail'] == 'Fail')
][info_columns + ['Cert_Pass_Fail', 'Real_Pass_Fail', 'Cert_Direction', 'Real_Direction']]

print("\nList of Items that Failed the Test:")
if show_full_lists:
    print(failed_items)
else:
    print(f"Total number of items that failed the test: {len(failed_items)}")

failed_items.to_excel(os.path.join(output_folder, 'Failed_Items.xlsx'), index=False)

df_matched['Percent_Difference'] = ((df_matched['Real_Mean_Test'] - df_matched['Cert_Mean_Test']) / df_matched['Cert_Mean_Test']) * 100
high_difference_samples = df_matched[df_matched['Percent_Difference'].abs() > 200][
    info_columns + ['Cert_Mean_Test', 'Real_Mean_Test', 'Percent_Difference', 'Cert_Direction', 'Real_Direction']
]

print("\nSamples with Percentage Difference Greater Than 200%:")
if show_full_lists:
    print(high_difference_samples)
else:
    print(f"Total number of samples with >200% difference: {len(high_difference_samples)}")

high_difference_samples.to_excel(os.path.join(output_folder, 'High_Percent_Difference_Samples.xlsx'), index=False)

print("Preparing model data...")
base_features = ['QC_Re', 'QC_Rm', 'QC_A', 'Dimension']
original_features = base_features + ['Manufacturer']
targets = ['Yield Strength [Mpa]', 'Tensile Strength [in MPA]', 'Elongation']

for col in base_features + targets + ['Elongation', 'QC_A']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Manufacturer'] = df['Manufacturer'].astype(str)
df_model = df.dropna(subset=original_features + targets).copy()

df_model = create_enhanced_features(df_model)
df_model = pd.get_dummies(df_model, columns=['Manufacturer'])

new_feature_list = [
    'Yield_Tensile_Ratio', 'Total_Elongation_Energy', 'Strength_Difference_Ratio',
    'Modulus_of_Toughness', 'Hardening_Coefficient', 'Elongation_Thickness_Ratio',
    'Reciprocal_Yield_Strength', 'Reciprocal_Tensile_Strength'
]

enhanced_feature_list = [
    'Yield_Ductility_Product', 'Work_Hardening_Range', 'Normalized_Work_Hardening',
    'Work_Hardening_Ductility', 'Ductility_Weighted_Strength', 'Quality_Index',
    'Performance_Index', 'Strain_Hardening_Coefficient', 'Strain_Energy',
    'Strength_Ductility_Balance', 'Formability_Index', 'Thickness_Strength_Index',
    'Thickness_Ductility_Ratio', 'Hall_Petch_Approximation', 'Size_Effect_Factor',
    'Thickness_Quality_Index', 'Thickness_Strain_Energy', 'Normalized_Thickness_Strength',
    'Thickness_Performance_Factor', 'Surface_Volume_Strength'
] + new_feature_list

manufacturer_cols = [col for col in df_model.columns if col.startswith('Manufacturer_')]
selected_features = base_features + enhanced_feature_list + manufacturer_cols

for target in targets:
    df_model[target] = pd.to_numeric(df_model[target], errors='coerce')

df_model = df_model.dropna(subset=selected_features + targets).copy()

# Outlier removal
clean_indices = df_model.index
for target in targets:
    y_z_scores = np.abs(zscore(df_model.loc[clean_indices, target]))
    threshold = Outlier_threshold
    target_outlier_mask = y_z_scores > threshold
    outlier_indices = df_model.loc[clean_indices].index[target_outlier_mask]
    clean_indices = clean_indices.difference(outlier_indices)

df_model_clean = df_model.loc[clean_indices].copy()
outliers = df_model.drop(index=clean_indices)
outliers.to_excel(os.path.join(output_folder, 'Removed_Target_Outliers.xlsx'), index=False)
print(f"\nNumber of outliers removed from target variables: {len(outliers)}")

if 'Manufacturer' not in df_model_clean.columns:
    df_model_clean['Manufacturer'] = df.loc[df_model_clean.index, 'Manufacturer']

def apply_requirements(row):
    requirements = get_mechanical_requirements(row['Simplified Grade (for further analysis)'], row['Dimension'])
    row['Required_Yield_Strength'] = requirements['min_yield_strength']
    row['Required_Tensile_Min'] = requirements['tensile_range'][0]
    row['Required_Tensile_Max'] = requirements['tensile_range'][1]
    row['Required_Elongation'] = requirements['min_elongation']
    return row

df_model_clean = df_model_clean.apply(apply_requirements, axis=1)
df_model_clean = df_model_clean.dropna(subset=['Yield Strength [Mpa]', 'Tensile Strength [in MPA]', 'Elongation', 'Required_Yield_Strength'])

def determine_pass_fail_actual(row):
    failures = []
    row['Actual_Pass_Fail_Yield Strength [Mpa]'] = 'Pass'
    row['Actual_Pass_Fail_Tensile Strength [in MPA]'] = 'Pass'
    row['Actual_Pass_Fail_Elongation'] = 'Pass'

    if row['Yield Strength [Mpa]'] < row['Required_Yield_Strength']:
        failures.append('Yield Strength')
        row['Actual_Pass_Fail_Yield Strength [Mpa]'] = 'Fail'

    if row['Elongation'] < row['Required_Elongation']:
        failures.append('Elongation')
        row['Actual_Pass_Fail_Elongation'] = 'Fail'

    if row['Tensile Strength [in MPA]'] < row['Required_Tensile_Min']:
        failures.append('Tensile Strength Below Min')
        row['Actual_Pass_Fail_Tensile Strength [in MPA]'] = 'Fail'

    if failures:
        row['Failure_Reason'] = ', '.join(sorted(failures))
        row['Actual_Pass_Fail'] = 'Fail'
    else:
        row['Failure_Reason'] = 'None'
        row['Actual_Pass_Fail'] = 'Pass'

    return row

df_model_clean = df_model_clean.apply(determine_pass_fail_actual, axis=1)

manufacturer_sample_counts = df_model_clean['Manufacturer'].value_counts().to_dict()

def determine_pass_fail_predicted(row, target_variable):
    manufacturer = row['Manufacturer']
    manufacturer_sample_size = manufacturer_sample_counts.get(manufacturer, 0)
    if manufacturer_sample_size < minimum_sample_size:
        return 'Fail'

    if target_variable == 'Yield Strength [Mpa]':
        return 'Fail' if row['Pred_Lower'] < row['Required_Yield_Strength'] else 'Pass'
    elif target_variable == 'Tensile Strength [in MPA]':
        return 'Fail' if row['Pred_Lower'] < row['Required_Tensile_Min'] else 'Pass'
    elif target_variable == 'Elongation':
        return 'Fail' if row['Pred_Lower'] < row['Required_Elongation'] else 'Pass'

def analyze_confusion_matrix(y_true, y_pred, target_name, grade):
    cm = confusion_matrix(y_true, y_pred, labels=['Pass', 'Fail'])
    if cm.shape == (2,2):
        tn, fp, fn, tp = cm.ravel()
    else:
        tn, fp, fn, tp = 0,0,0,0

    if (tn+fp+fn+tp) > 0:
        accuracy_val = accuracy_score(y_true, y_pred)*100
        precision_val = precision_score(y_true, y_pred, pos_label='Fail')*100 if ('Fail' in y_pred.values) else np.nan
        recall_val = recall_score(y_true, y_pred, pos_label='Fail')*100 if ('Fail' in y_true.values) else np.nan
        f1_val = f1_score(y_true, y_pred, pos_label='Fail')*100 if ('Fail' in y_true.values and 'Fail' in y_pred.values) else np.nan
    else:
        accuracy_val, precision_val, recall_val, f1_val = np.nan, np.nan, np.nan, np.nan

    print(f"\nConfusion Matrix Analysis for {target_name} (Grade: {grade})")
    print("-" * 50)
    print(f"TN: {tn}, FP: {fp}, FN: {fn}, TP: {tp}")
    print("\nMetrics:")
    if not np.isnan(accuracy_val):
        print(f"Accuracy: {accuracy_val:.2f}%")
    if not np.isnan(precision_val):
        print(f"Precision: {precision_val:.2f}%")
    if not np.isnan(recall_val):
        print(f"Recall: {recall_val:.2f}%")
    if not np.isnan(f1_val):
        print(f"F1 Score: {f1_val:.2f}%")

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pass', 'Fail'], yticklabels=['Pass', 'Fail'])
    plt.title(f'Confusion Matrix for {target_name} - Grade: {grade}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f"Confusion_Matrix_{target_name}_{grade}.png"))
    plt.show()

    return {
        'tn': tn,
        'fp': fp,
        'fn': fn,
        'tp': tp,
        'accuracy': accuracy_val,
        'precision': precision_val,
        'recall': recall_val,
        'f1': f1_val
    }

def model_and_evaluate_ensemble(df_subset):
    results = {}
    X = df_subset[selected_features]

    for target_variable in targets:
        print(f"\n{'='*40}\nPredicting {target_variable} for subset grade with Ensemble\n{'='*40}")
        y = df_subset[target_variable]

        rf = RandomForestRegressor(
            n_estimators=rf_n_estimators,
            max_depth=rf_max_depth,
            min_samples_split=rf_min_samples_split,
            min_samples_leaf=rf_min_samples_leaf,
            max_features=rf_max_features,
            bootstrap=rf_bootstrap,
            oob_score=rf_oob_score,
            random_state=random_state
        )

        et = ExtraTreesRegressor(
            n_estimators=et_n_estimators,
            max_depth=et_max_depth,
            min_samples_split=et_min_samples_split,
            min_samples_leaf=et_min_samples_leaf,
            max_features=et_max_features,
            random_state=random_state
        )

        models = [rf, et]
        model_names = ['Random Forest', 'Extra Trees']

        # Single random state for simplicity
        rs = random_state
        kf = KFold(n_splits=num_cv_splits, shuffle=True, random_state=rs)

        cv_predictions = {name: [] for name in model_names}
        cv_y_tests = []
        cv_indices = []

        cv_lowers = {name: [] for name in model_names}
        cv_uppers = {name: [] for name in model_names}

        fi_list = []

        for fold, (train_index, test_index) in enumerate(kf.split(X)):
            print(f"Fold {fold + 1}/{num_cv_splits} RS {rs}")
            X_train_cv, X_test_cv = X.iloc[train_index], X.iloc[test_index]
            y_train_cv, y_test_cv = y.iloc[train_index], y.iloc[test_index]
            cv_y_tests.append(y_test_cv)
            cv_indices.append(test_index)

            fold_fi_list = []
            for model, name in zip(models, model_names):
                model.fit(X_train_cv, y_train_cv)
                y_pred_cv = model.predict(X_test_cv)
                cv_predictions[name].append(y_pred_cv)

                rmse_cv = np.sqrt(mean_squared_error(y_test_cv, y_pred_cv))
                r2_cv = r2_score(y_test_cv, y_pred_cv)
                print(f"{name} - Fold {fold + 1} RMSE: {rmse_cv:.4f}, R²: {r2_cv:.4f}")

                lower, upper = calculate_prediction_intervals_for_tree_model(model, X_test_cv, confidence_level)
                cv_lowers[name].append(lower)
                cv_uppers[name].append(upper)

                if hasattr(model, 'feature_importances_'):
                    fold_fi_list.append(model.feature_importances_)

            if fold_fi_list:
                fi_fold_avg = np.mean(fold_fi_list, axis=0)
                fi_list.append(fi_fold_avg)

        combined_y_test = pd.concat(cv_y_tests)
        combined_indices = np.concatenate(cv_indices)
        sorted_indices = np.argsort(combined_indices)
        combined_y_test = combined_y_test.iloc[sorted_indices]

        combined_predictions = {}
        combined_lowers = {}
        combined_uppers = {}
        for name in model_names:
            combined_pred = np.concatenate(cv_predictions[name])[sorted_indices]
            combined_predictions[name] = combined_pred

            combined_lower = np.concatenate(cv_lowers[name])[sorted_indices]
            combined_upper = np.concatenate(cv_uppers[name])[sorted_indices]
            combined_lowers[name] = combined_lower
            combined_uppers[name] = combined_upper

        # Ensemble predictions
        ensemble_predictions = np.mean([combined_predictions[name] for name in model_names], axis=0)
        y_pred_lower_ensemble = np.mean([combined_lowers[name] for name in model_names], axis=0)
        y_pred_upper_ensemble = np.mean([combined_uppers[name] for name in model_names], axis=0)

        combined_indices = combined_indices[sorted_indices]

        mse = mean_squared_error(combined_y_test, ensemble_predictions)
        rmse = np.sqrt(mse)
        r2 = r2_score(combined_y_test, ensemble_predictions)

        print(f"\nEnsemble Model Performance for {target_variable}:")
        print(f"Combined Test RMSE: {rmse:.2f}")
        print(f"Combined Test R²: {r2:.4f}")

        X_test_with_index = X.iloc[combined_indices].copy()
        X_test_with_index[f'{target_variable}_Actual'] = combined_y_test.values
        X_test_with_index[f'{target_variable}_Predicted'] = ensemble_predictions
        X_test_with_index['Pred_Lower'] = y_pred_lower_ensemble
        X_test_with_index['Pred_Upper'] = y_pred_upper_ensemble

        if fi_list:
            avg_importances = np.mean(fi_list, axis=0)
            feature_names = X.columns
            feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': avg_importances})
            feature_importances.sort_values(by='Importance', ascending=False, inplace=True)

            plt.figure(figsize=(12, 8))
            sns.barplot(x='Importance', y='Feature', data=feature_importances.head(20))
            plt.title(f'Top 20 Feature Importances for {target_variable}')
            plt.tight_layout()
            plt.savefig(os.path.join(output_folder, f"Feature_Importance_{target_variable}.png"))
            plt.show()

        certificate_columns = {
            'Yield Strength [Mpa]': 'QC_Re',
            'Tensile Strength [in MPA]': 'QC_Rm',
            'Elongation': 'QC_A'
        }
        certificate_column = certificate_columns.get(target_variable)
        if certificate_column:
            X_test_with_index[f'{target_variable}_Certificate'] = df_subset.loc[X_test_with_index.index, certificate_column]
        else:
            X_test_with_index[f'{target_variable}_Certificate'] = np.nan

        results[target_variable] = {
            'models': [rf, et],
            'X_test_with_index': X_test_with_index,
            'y_test': combined_y_test,
            'y_pred': ensemble_predictions,
            'y_pred_lower': y_pred_lower_ensemble,
            'y_pred_upper': y_pred_upper_ensemble
        }

    return results

grades_to_analyze = ['S690', 'S355', 'Q345']
df_model_clean_grades = df_model_clean[df_model_clean['Simplified Grade (for further analysis)'].isin(grades_to_analyze)]

grade_results = {}
for grade in grades_to_analyze:
    df_subset = df_model_clean_grades[df_model_clean_grades['Simplified Grade (for further analysis)'] == grade].copy()
    if df_subset.empty:
        print(f"No data for grade {grade}")
        continue
    if len(df_subset) < 100:
        print(f"Warning: Not enough samples for grade {grade}")
    print(f"\n\n***** Processing Grade: {grade} *****\n\n")
    grade_results[grade] = model_and_evaluate_ensemble(df_subset)

# Plotting and Analysis per Grade
for grade in grade_results:
    result = grade_results[grade]
    df_subset = df_model_clean_grades[df_model_clean_grades['Simplified Grade (for further analysis)'] == grade].copy()
    overall_results = pd.DataFrame(index=result[targets[0]]['X_test_with_index'].index)
    overall_results['Actual_Pass_Fail'] = df_subset.loc[overall_results.index, 'Actual_Pass_Fail']

    for target in targets:
        X_test = result[target]['X_test_with_index']
        X_test['Dimension'] = df_subset.loc[X_test.index, 'Dimension']
        X_test['Required_Yield_Strength'] = df_subset.loc[X_test.index, 'Required_Yield_Strength']
        X_test['Required_Tensile_Min'] = df_subset.loc[X_test.index, 'Required_Tensile_Min']
        X_test['Required_Tensile_Max'] = df_subset.loc[X_test.index, 'Required_Tensile_Max']
        X_test['Required_Elongation'] = df_subset.loc[X_test.index, 'Required_Elongation']
        X_test['Actual_Pass_Fail'] = df_subset.loc[X_test.index, f'Actual_Pass_Fail_{target}']
        X_test['Failure_Reason'] = df_subset.loc[X_test.index, 'Failure_Reason']
        X_test['Manufacturer'] = df_subset.loc[X_test.index, 'Manufacturer']
        X_test['Predicted_Pass_Fail'] = X_test.apply(lambda row: determine_pass_fail_predicted(row, target), axis=1)

        overall_results[f'Predicted_Pass_Fail_{target}'] = X_test['Predicted_Pass_Fail']

        y_true = X_test['Actual_Pass_Fail']
        y_pred = X_test['Predicted_Pass_Fail']
        analyze_confusion_matrix(y_true, y_pred, f"{target}", grade)

        combined_results = X_test.copy()
        color_map = {'Pass': 'blue', 'Fail': 'orange'}
        combined_results['Color'] = combined_results['Actual_Pass_Fail'].map(color_map)
        combined_results = combined_results.dropna(subset=['Color'])
        combined_results.sort_values(by='Dimension', inplace=True)

        # Plot predicted vs actual for this target and grade
        plt.figure(figsize=(8,6))
        # Ensure numeric
        actual_vals = pd.to_numeric(combined_results[f'{target}_Actual'], errors='coerce')
        pred_vals = pd.to_numeric(combined_results[f'{target}_Predicted'], errors='coerce')
        mask = pd.notna(actual_vals) & pd.notna(pred_vals)
        actual_vals = actual_vals[mask]
        pred_vals = pred_vals[mask]

        plt.scatter(actual_vals, pred_vals, alpha=0.5)
        min_val = min(actual_vals.min(), pred_vals.min())
        max_val = max(actual_vals.max(), pred_vals.max())
        plt.plot([min_val, max_val],[min_val, max_val],'r--', label='Perfect Fit')
        plt.xlabel('Actual')
        plt.ylabel('Predicted')
        plt.title(f'{target} Actual vs Predicted - Grade {grade}')
        plt.legend()
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, f'Actual_vs_Predicted_{target}_{grade}.png'))
        plt.show()

        # Compute RMSE and MAE per manufacturer for this target and grade
        manufacturer_metrics = combined_results.groupby('Manufacturer').apply(lambda g: pd.Series({
            'RMSE': np.sqrt(mean_squared_error(g[f'{target}_Actual'], g[f'{target}_Predicted'])),
            'MAE': np.mean(np.abs(g[f'{target}_Actual'] - g[f'{target}_Predicted']))
        }))

        # Plot RMSE and MAE per manufacturer
        for metric in ['RMSE', 'MAE']:
            plt.figure(figsize=(10,6))
            manufacturer_metrics.sort_values(by=metric, ascending=True, inplace=True)
            sns.barplot(x=metric, y=manufacturer_metrics.index, data=manufacturer_metrics.reset_index(), palette='viridis')
            plt.title(f'{metric} per Manufacturer for {target} - Grade {grade}')
            plt.xlabel(metric)
            plt.ylabel('Manufacturer')
            plt.tight_layout()
            plt.savefig(os.path.join(output_folder, f'{metric}_per_Manufacturer_{target}_{grade}.png'))
            plt.show()

    def determine_overall_predicted_pass_fail(row):
        return 'Pass' if all(row[f'Predicted_Pass_Fail_{t}'] == 'Pass' for t in targets) else 'Fail'
    overall_results['Predicted_Pass_Fail_Overall'] = overall_results.apply(determine_overall_predicted_pass_fail, axis=1)

    y_true_overall = overall_results['Actual_Pass_Fail']
    y_pred_overall = overall_results['Predicted_Pass_Fail_Overall']
    analyze_confusion_matrix(y_true_overall, y_pred_overall, 'Overall Requirements', grade)

    overall_results.to_excel(os.path.join(output_folder, f'Model_Results_{grade}.xlsx'), index=False)

    # Best and Worst Manufacturers by passing rate are already computed in manufacturer_stats
    manufacturer_groups = df_model_clean.groupby('Manufacturer')
    manufacturer_stats = manufacturer_groups.apply(
        lambda group: pd.Series({
            'Mean_Certificate_Yield_Strength': group['QC_Re'].mean(),
            'Mean_Actual_Yield_Strength': group['Yield Strength [Mpa]'].mean(),
            'Mean_Yield_Strength_Difference': group['Yield Strength [Mpa]'].mean() - group['QC_Re'].mean(),
            'Std_Yield_Strength': group['Yield Strength [Mpa]'].std(),
            'Sample_Size': len(group),
            'Pass_Count': (group['Actual_Pass_Fail'] == 'Pass').sum(),
            'Fail_Count': (group['Actual_Pass_Fail'] == 'Fail').sum(),
            'Pass_Rate': (group['Actual_Pass_Fail'] == 'Pass').mean() * 100
        })
    ).reset_index()

    manufacturer_stats_filtered = manufacturer_stats[manufacturer_stats['Sample_Size'] >= minimum_sample_size]

    top_manufacturers = manufacturer_stats_filtered.sort_values(by='Pass_Rate', ascending=False).head(1)
    worst_manufacturers = manufacturer_stats_filtered.sort_values(by='Pass_Rate', ascending=True).head(1)

    selected_manufacturers = pd.concat([top_manufacturers['Manufacturer'], worst_manufacturers['Manufacturer']])

    for manufacturer in selected_manufacturers:
        group_data = df_model_clean[df_model_clean['Manufacturer'] == manufacturer]['Yield Strength [Mpa]']

        mean_yield = group_data.mean()
        std_yield = group_data.std()

        plt.figure(figsize=(10, 6))
        sns.histplot(group_data, kde=True, bins=30, color='skyblue', edgecolor='black')
        plt.axvline(mean_yield, color='red', linestyle='--', label='Mean Yield Strength')
        plt.axvline(mean_yield + std_yield, color='green', linestyle='--', label='+1 Std Dev')
        plt.axvline(mean_yield - std_yield, color='green', linestyle='--', label='-1 Std Dev')
        plt.axvline(mean_yield + 3 * std_yield, color='purple', linestyle='--', label='+3 Std Dev')
        plt.axvline(mean_yield - 3 * std_yield, color='purple', linestyle='--', label='-3 Std Dev')

        ci_lower = mean_yield - 3 * std_yield
        ci_upper = mean_yield + 3 * std_yield
        plt.fill_betweenx([0, plt.gca().get_ylim()[1]], ci_lower, ci_upper, color='yellow', alpha=0.2, label='99.7% CI')

        plt.title(f'Yield Strength Distribution for {manufacturer}')
        plt.xlabel('Yield Strength [Mpa]')
        plt.ylabel('Frequency')
        plt.legend()
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, f'Yield_Strength_Distribution_{manufacturer}.png'))
        plt.show()

    failed_samples = df_model_clean[df_model_clean['Actual_Pass_Fail'] == 'Fail']
    failure_reasons_series = failed_samples['Failure_Reason'].str.split(', ', expand=True).stack()
    failure_reason_counts = failure_reasons_series.value_counts()
    failure_reason_percentages = (failure_reason_counts / failure_reason_counts.sum()) * 100

    failure_reason_df = pd.DataFrame({
        'Failure Reason': failure_reason_percentages.index,
        'Percentage': failure_reason_percentages.values
    })

    plt.figure(figsize=(10, 6))
    sns.barplot(
        x='Percentage',
        y='Failure Reason',
        data=failure_reason_df,
        palette='viridis'
    )
    plt.title('Percentage of Failures per Failure Reason', fontsize=16)
    plt.xlabel('Percentage (%)', fontsize=14)
    plt.ylabel('Failure Reason', fontsize=14)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'Failure_Reason_Percentage_{grade}.png'))
    plt.show()

    false_negatives = overall_results[
        (overall_results['Actual_Pass_Fail'] == 'Fail') &
        (overall_results['Predicted_Pass_Fail_Overall'] == 'Pass')
    ]

    false_negative_samples = df_model_clean.loc[false_negatives.index].copy()
    false_negative_samples['Predicted_Pass_Fail_Overall'] = overall_results.loc[false_negatives.index, 'Predicted_Pass_Fail_Overall']

    false_negative_list = false_negative_samples[
        ['Material Code', 'Dimension', 'Yield Strength [Mpa]', 'Elongation', 'Tensile Strength [in MPA]',
         'Required_Yield_Strength', 'Required_Elongation', 'Required_Tensile_Min', 'Required_Tensile_Max',
         'Failure_Reason']
    ]

    print("\nList of False Negatives (Actual Failures Predicted as Pass):")
    if show_full_lists:
        print(false_negative_list)
    else:
        print(f"Total number of false negatives: {len(false_negative_list)}")

    false_negative_list.to_excel(os.path.join(output_folder, f'False_Negatives_List_{grade}.xlsx'), index=False)

    # Plot Pass Rates for all manufacturers
    plt.figure(figsize=(10,6))
    sorted_manuf = manufacturer_stats_filtered.sort_values(by='Pass_Rate', ascending=False)
    sns.barplot(x='Pass_Rate', y='Manufacturer', data=sorted_manuf, palette='coolwarm')
    plt.title(f'Passing Rate per Manufacturer - Grade {grade}')
    plt.xlabel('Pass Rate (%)')
    plt.ylabel('Manufacturer')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'Pass_Rate_per_Manufacturer_{grade}.png'))
    plt.show()

print("All steps completed successfully!")

# =============================================
# PREDICT SINGLE SAMPLE FUNCTION
# =============================================
def predict_single_sample(grade, dimension, manufacturer, QC_Re, QC_Rm, QC_A):
    if grade not in grade_results:
        raise ValueError(f"No trained models for grade {grade}")

    df_input = pd.DataFrame({
        'QC_Re': [QC_Re],
        'QC_Rm': [QC_Rm],
        'QC_A': [QC_A],
        'Dimension': [dimension],
        'Manufacturer': [manufacturer]
    })

    df_input = create_enhanced_features(df_input)

    for col in manufacturer_cols:
        df_input[col] = 0
    manu_col = f'Manufacturer_{manufacturer}'
    if manu_col in df_input.columns:
        df_input[manu_col] = 1

    X_sample = df_input[selected_features]

    reqs = get_mechanical_requirements(grade, dimension)
    required_yield = reqs['min_yield_strength']
    required_tensile_min = reqs['tensile_range'][0]
    required_tensile_max = reqs['tensile_range'][1]
    required_elongation = reqs['min_elongation']

    predictions = {}
    intervals = {}
    pass_fail = {}

    for target in targets:
        models = grade_results[grade][target]['models']

        model_preds = []
        model_lower = []
        model_upper = []
        for model in models:
            pred = model.predict(X_sample)[0]
            if hasattr(model, 'estimators_'):
                all_tree_preds = np.array([est.predict(X_sample)[0] for est in model.estimators_])
                alpha = (100 - confidence_level) / 2
                lower_percentile = alpha
                upper_percentile = 100 - alpha
                lower_val = np.percentile(all_tree_preds, lower_percentile)
                upper_val = np.percentile(all_tree_preds, upper_percentile)
            else:
                lower_val = pred - Z_value * 10
                upper_val = pred + Z_value * 10

            model_preds.append(pred)
            model_lower.append(lower_val)
            model_upper.append(upper_val)

        final_pred = np.mean(model_preds)
        final_lower = np.mean(model_lower)
        final_upper = np.mean(model_upper)

        predictions[target] = final_pred
        intervals[target] = (final_lower, final_upper)

        if target == 'Yield Strength [Mpa]':
            pf = 'Fail' if final_lower < required_yield else 'Pass'
        elif target == 'Tensile Strength [in MPA]':
            pf = 'Fail' if final_lower < required_tensile_min else 'Pass'
        elif target == 'Elongation':
            pf = 'Fail' if final_lower < required_elongation else 'Pass'

        pass_fail[target] = pf

    overall_pass_fail = 'Pass' if all(p == 'Pass' for p in pass_fail.values()) else 'Fail'

    return {
        'Predictions': predictions,
        'Intervals': intervals,
        'Pass_Fail': pass_fail,
        'Overall_Pass_Fail': overall_pass_fail
    }
```
